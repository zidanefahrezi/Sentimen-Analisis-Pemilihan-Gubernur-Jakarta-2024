```python
import os

name = "All_dataset"

# Tentukan nama folder yang ingin dibuat
nama_folder = "/content/"+name+"/"

# Buat folder baru
os.makedirs(nama_folder, exist_ok=True)

print(f"Folder '{nama_folder}' berhasil dibuat.")
```

```python
import pandas as pd
import os

# Folder tempat file CSV berada
folder_path = nama_folder

# Membaca semua file CSV dalam folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Membaca dan menggabungkan semua file CSV
dataframes = []
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    dataframes.append(df)

# Menggabungkan semua DataFrame menjadi satu
combined_df = pd.concat(dataframes, ignore_index=True)

# Menyimpan dataset gabungan ke file baru
output_file = "/content/"+name+"_dataset.csv"
combined_df.to_csv(output_file, index=False)

print(f"Dataset berhasil digabungkan dan disimpan ke {output_file}")
```

```python
import pandas as pd

# Contoh DataFrame

df = pd.read_csv(output_file)

# Menghapus duplikat berdasarkan kolom 'Nama'
df_unique = df.drop_duplicates(subset=['full_text'])

print("DataFrame tanpa duplikat berdasarkan kolom 'Nama':")
df_unique.to_csv("/content/"+name+"_duplicate_dataset.csv", index=False)
```
